## Предобработка данных

### Изменение формата признаков

Во-первых, нужно привести названия столбцов к нужному формату.

```python
data.columns = data.columns.str.lower()
```

### Изменение типа данных в столбцах

Далее будем заменять тип данных в столбцах. Начнем с `user_score`.

Посмотрим уникальные значения столбца.

```python
data['user_score'].unique()
```

 На игровом рынке `tbd` значит "подлежит определению". Это значение вполне может заменить на `nan`.

```python
data['user_score'] = data['user_score'].replace('tbd', float('Nan'))
```

Теперь можно поменять тип данных в этом столбце.

```python
data['user_score'] = data['user_score'].astype('float')
```

Теперь перейдем к столбцу 'year_of_release'.

Посмотрим уникальные значения столбца.

```python
data['year_of_release'].unique()
```

Переведем тип данных в `int`, при этом используем метод, который применим для данных с пропусками.

```python
data['year_of_release'] = data['year_of_release'].round().astype('Int32', errors='ignore')
```

Проверим методом `info`.

```python
data.info()
```

Теперь каждому столбцу соответствует необходимый тип данных.

### Заполнение пропусков

```python
pd.DataFrame(round(data.isna().mean()*100,1)).style.background_gradient('coolwarm')
```

Больше всего пропусков предусмотрено в трех последних столбцах - `critic_score`, `user_score`, `rating`. Также есть небольшое количество пропусков в `year_of_release`.

```python
data['rating'].isna().sum()
```

В `critic_score` и `user_score` лучше не заменять пропуски средним значением и оставить как есть, иначе могут появиться искажения в данных, от самих строк также избавляться не будем, потому что они предусмотрены в существенном количестве. 

Что касается `rating`, пропуски там связаны с тем, что рейтинги выставлются американским агенством, не затрагивая игры в других регионах. Пропуски в этом признаке заполним значением "заглушкой" - `unknown`.

```python
data['rating'] = data['rating'].fillna('unknown')
```

Проверим.

```python
data['rating'].isna().sum()
```

Пропуски в столбце рейтинга заполнены.

Также есть пропуски в столбце года выпуска.

```python
data['year_of_release'].isna().sum()
```

Соотношение пропусков составляет менее 2%. Будем сбрасывать.

```python
data = data.dropna(subset=['year_of_release'])
```

Проверяем.

```python
data['year_of_release'].isna().sum()
```

Необходимые пропуски заполнены.

### Работа с дубликатами

Теперь проверим код на наличие явных дубликатов. Посмотрим по трем столбцам.

```python
print(data[['platform', 'name', 'year_of_release']].duplicated().sum())
```

Предусмотрено 2 дубликата, избавимся от них.

```python
data = data.drop_duplicates(subset=['platform', 'name', 'year_of_release'])
```

Проверим.

```python
data[['platform', 'name', 'year_of_release']].duplicated().sum()
```

Работа с дубликатами завершена.

### Feature Engineering

Создадим новый столбец для общих продаж.

```python
data['sum_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis='columns')
data['sum_sales'].head()
```

В этом столбце будут продажи по всем регионам.

Далее, переходим к исследовательскому анализу данных.

## Исследовательский анализ данных

### Рассмотрим распределение продаж по годам для разных платформ

Посмотрим сколько игр выпускалось в разные годы.

```python
games_per_year = (
    data
    .drop_duplicates(subset=['name', 'year_of_release'])
    .groupby('year_of_release')['name']
    .count()
)

games_per_year.plot(kind='bar', figsize=(12, 5))
plt.title('Количество выпущенных игр по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')
plt.show()
```

Количество выпускаемых игр постепенно росло с начала 1990-х годов и достигло максимума в конце 2000-х годов, с пиком около 2007 года. После этого наблюдается заметное снижение числа выпускаемых игр. В связи с этим для дальнейшего анализа и прогнозирования целесообразно рассматривать данные, начиная с 2000-х годов, так как они наиболее репрезентативны для текущего состояния рынка.

Далее, посмотрим продажи по платформам. Рассмотрим самые продаваемые за всю историю.

```python
plat_best_sales = data.groupby(['platform'])['sum_sales'].sum().sort_values(ascending=False)
plat_best_sales.head(10)
```

Наибольшие продажи за всю историю были у PS2.

Теперь построим для первых шести из вышеперечисленных платформ распределение продаж по годам.

```python
#plats_best_list = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS']
#plat_best_sales_data = data.query('platform == ["PS2", "X360", "PS3", "Wii", "DS", "PS"]')
```

```python
plats_best_list = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS']
plat_best_sales_data = data[data['platform'].isin(plats_best_list)]
```

Построим общий график.

```python
sales_years = plat_best_sales_data.pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum',  fill_value=0)
```

```python
sales_years = sales_years.sort_index()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sales_years.plot(
    ax=ax,
    linewidth=2,
    alpha=0.85
)

ax.set_title('Динамика продаж по платформам', fontsize=18)
ax.set_xlabel('Год выпуска')
ax.set_ylabel('Продажи, млн копий')
ax.set_xlim(1993, 2016)

ax.legend(
    title='Платформа',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

ax.grid(True, alpha=0.3)
plt.show()
```

Рассмотрим среднее время жизни платформ.

```python
platform_life = (
    data.groupby('platform')['year_of_release']
    .agg(['min', 'max'])
)

platform_life['lifetime'] = platform_life['max'] - platform_life['min']
```

```python
platform_life['lifetime'].describe()
```

```python
print(platform_life['lifetime'].median())
```

```python
platform_life['lifetime'].plot(
    kind='box',
    figsize=(8, 2),
    vert=False
)

plt.title('Срок жизни игровых платформ')
plt.xlabel('Лет')
plt.show()
```

В данных присутствуют выбросы — платформы с существенно более долгим жизненным циклом(PC и DS), однако они не являются типичными.

```python
mean_life = platform_life['lifetime'].mean()
median_life = platform_life['lifetime'].median()

platform_life['lifetime'].plot(
    kind='hist',
    bins=10,
    figsize=(9, 4),
    alpha=0.8
)

plt.axvline(mean_life, color='red', linestyle='--', label=f'Среднее: {mean_life:.1f}')
plt.axvline(median_life, color='black', linestyle=':', label=f'Медиана: {median_life:.1f}')

plt.title('Распределение срока жизни игровых платформ')
plt.xlabel('Срок жизни, лет')
plt.ylabel('Количество платформ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Средний срок жизни игровой платформы составляет около 7–8 лет, медианное значение близко к этому показателю. Большинство платформ существует в пределах одного десятилетия, после чего их продажи значительно снижаются и на рынок выходят новые поколения устройств. Это подтверждает наличие выраженного жизненного цикла у игровых платформ. Рассматривая, например, линейку консолей PlayStation, можно сделать вывод, что промежуток времени между обновлением поколений примерно соответствует этой цифре. PS (1994) - PS2 (2000) - PS3 (2006) - PS4 (2013), а PS5 вышла в 2020.

### Распределение продаж для актуальных данных

В результате анализа динамики выпуска игр и жизненного цикла платформ в качестве актуального периода были выбраны данные, начиная с 2014 года. Этот период соответствует началу текущего поколения консолей и наиболее отражает современное состояние рынка.

Оформим соответствующий срез в новую переменную и посмотрим наиболее продаваемые платформы за актуальный период.

```python
good_data = data[data['year_of_release'] >= 2014]
good_data.groupby(['platform'])['sum_sales'].sum().sort_values(ascending=False)
```

Список существенно поменялся. Рассмотрим графики продаж актуальных платформ для актуального промежутка времени.

```python
sales_actual = good_data.pivot_table(
    index='year_of_release',
    columns='platform',
    values='sum_sales',
    aggfunc='sum',
    fill_value=0
).sort_index()

ax = sales_actual.plot(figsize=(10, 6), linewidth=2)

ax.set_title('Динамика продаж платформ (2014–2016)')
ax.set_xlabel('Год выпуска')
ax.set_ylabel('Продажи, млн копий')

ax.set_xticks([2014, 2015, 2016])
ax.set_xlim(2013.8, 2016.2)

ax.grid(True, alpha=0.3)
plt.show()
```

Рассмотрим платформы.

В актуальном периоде 2014–2016 годов лидерами рынка являются платформы нового поколения — PS4 и Xbox One. PS4 демонстрирует наибольшие объёмы продаж на протяжении всего периода, несмотря на снижение показателей в 2016 году, что может быть связано с неполнотой данных за последний год. Xbox One занимает второе место и показывает сопоставимую динамику, уступая PS4 по объёму продаж, но значительно опережая остальные платформы.

Платформы предыдущего поколения (PS3, Xbox 360, Wii, Wii U) демонстрируют устойчивое снижение продаж, что указывает на завершение их жизненного цикла в рамках рассматриваемого периода. Платформа 3DS сохраняет определённый уровень спроса, однако её продажи также снижаются.

Платформы PC и PSV характеризуются сравнительно низкими объёмами продаж на протяжении всего актуального периода и не могут рассматриваться как ключевые драйверы рынка. В целом, на основе динамики продаж наиболее перспективными платформами для дальнейшего анализа являются PS4 и Xbox One.

Построим "Ящик с усами" для вышеперечисленных платформ и убедимся в выводах.

```python
plat_sales_years = good_data.query("platform == ['PS4', 'XOne', 'PC', 'WiiU', 'PSV', '3DS', 'X360', 'PS3']").\
pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
plat_sales_years.sort_values(['year_of_release'], ascending=False)
```

```python
good_data.boxplot(column='sum_sales', by='platform', figsize=(10, 10));
plt.ylim([0, 3])
```

Диаграмма размаха показывает, что наибольший разброс продаж характерен для платформ PS4 и Xbox One, что указывает на наличие значительного количества игр с высокими продажами.
```python
plat_user = good_data.pivot_table(index='platform', values='user_score')
plat_user.sort_values(by='user_score', ascending=False)
```

Самые лучшие отзывы игроков - на портативных консолях. Потом уже идут консоли нового поколения.

Рассмотрим для критиков.

```python
plat_critic = good_data.pivot_table(index='platform', values='critic_score')
plat_critic.sort_values(by='critic_score', ascending=False)
```

Самые лучшие отзывы критиков - на PC, потом идут консоли нового поколения, потом портативные.

Также можно сделать вывод, что самые худшие отзывы в обоих категориях - на консолях старого поколения (PS3 и XBOX360).

Построим диаграмму рассеяния продаж и оценок и выведем коэффициент их корреляции для каждой платформы. Поочередно для каждой платформы - сначала для оценок критиков, потом пользователей.

```python
import matplotlib.pyplot as plt

platforms = good_data['platform'].dropna().unique()

for plat in platforms:
    df = good_data[good_data['platform'] == plat]

    # если мало данных — пропускаем, чтобы не плодить пустые графики
    if df.shape[0] < 10:
        continue

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    # 1) Critic score vs sales
    d1 = df[['critic_score', 'sum_sales']].dropna()
    if len(d1) >= 3:
        ax[0].scatter(d1['critic_score'], d1['sum_sales'], s=15, alpha=0.6)
        corr_critic = d1['sum_sales'].corr(d1['critic_score'])
    else:
        corr_critic = float('nan')
    ax[0].set_title(f'{plat}: Critic vs Sales (corr={corr_critic:.2f})')
    ax[0].set_xlabel('critic_score')
    ax[0].set_ylabel('sum_sales')

    # 2) User score vs sales
    d2 = df[['user_score', 'sum_sales']].dropna()
    if len(d2) >= 3:
        ax[1].scatter(d2['user_score'], d2['sum_sales'], s=15, alpha=0.6)
        corr_user = d2['sum_sales'].corr(d2['user_score'])
    else:
        corr_user = float('nan')
    ax[1].set_title(f'{plat}: User vs Sales (corr={corr_user:.2f})')
    ax[1].set_xlabel('user_score')
    ax[1].set_ylabel('sum_sales')

    plt.suptitle(f'Платформа {plat}', y=1.05, fontsize=14)
    plt.tight_layout()
    plt.show()

    print(plat, '| corr_critic =', corr_critic, '| corr_user =', corr_user)
```

Можем видеть, что для консолей старого и нового поколения отзывы критиков имеют серьезное значение для продаж (в отличие от отзывов пользователей). 

А для портативных консолей - WiiU и 3DS, значение имеет обе категории отзывов (PSV в эту категорию не относится).

Продажи на PC и PSV мало зависят от каких-либо отзывов.

Далее, для каждого региона рассмотрим самые популярные платформы, жанры и влияние возрастного рейтинга.

```python
import matplotlib.pyplot as plt

def region_profile(data, region_sales, region_name):
    """
    Портрет пользователя по региону:
    - таблицы groupby
    - графики (bar / pie)
    """

    platforms = (
        data.groupby('platform')[region_sales]
        .sum()
        .sort_values(ascending=False)
    )

    print(f'\n{region_name}. Продажи по платформам')
    display(platforms)

    platforms.plot(
        kind='bar',
        figsize=(6, 4),
        title=f'{region_name}: продажи по платформам',
        ylabel='Продажи, млн копий',
        xlabel='Платформа',
        rot=45
    )
    plt.show()

    genres = (
        data.groupby('genre')[region_sales]
        .sum()
        .sort_values(ascending=False)
    )

    print(f'\n{region_name}. Продажи по жанрам')
    display(genres)

    genres.plot(
        kind='bar',
        figsize=(6, 4),
        title=f'{region_name}: продажи по жанрам',
        ylabel='Продажи, млн копий',
        xlabel='Жанр',
        rot=45
    )
    plt.show()


    ratings = (
        data.groupby('rating')[region_sales]
        .sum()
        .sort_values(ascending=False)
    )

    print(f'\n{region_name}. Продажи по рейтингу ESRB')
    display(ratings)

    ratings.plot(
        kind='pie',
        figsize=(5, 5),
        autopct='%1.1f%%',
        startangle=90,
        title=f'{region_name}: продажи по рейтингу ESRB'
    )
    plt.ylabel('')
    plt.show()


```

H₀ (нулевая гипотеза): средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.

H₁ (альтернативная гипотеза): средние пользовательские рейтинги платформ Xbox One и PC различаются.

Уровень значимости: `α` = 0.05


Первоначально необходимо взглянуть на средние пользовательские оценки.

```python
avg_plat_user_scores = data_for_hip.groupby('platform')['user_score'].mean()
avg_plat_user_scores 
```

Сделаем двухсторонний тест, потому что нам не важно, в какую сторону значения отличаются. Уровень значимости - 5%.

```python
pc_data = data_for_hip.query('platform == "PC"')
xone_data = data_for_hip.query('platform == "XOne"')
```

```python
sample_1 = pc_data['user_score']
sample_2 = xone_data['user_score']

alpha = .05 

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

### Проверка второй гипотезы

Теперь рассмотрим вторую гипотезу: `средние пользовательские рейтинги жанров Action и Sports разные`:

H₀ (нулевая гипотеза): средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.

H₁ (альтернативная гипотеза): средние пользовательские рейтинги жанров Action и Sports различаются.

Уровень значимости: α = 0.05

Полученные результаты позволяют выделить перспективные платформы и жанры, а также учитывать региональные особенности рынка, что может быть использовано интернет-магазином при планировании дальнейших действий и рекламных кампаний.
